```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
```

### Method 1

```python
def exp_neg(x, N):
    value = 0
    for i in range(N+1):
        value = value +  (-x)**i/factorial(i)
    return value
```

```python
x=1
results = np.zeros(100)
Ns = np.arange(1,101)

for N in Ns:
    results[N-1] = exp_neg(x, N)
```

```python
fig, ax = plt.subplots()
ax.set(
    xlabel="Number of terms N",
    ylabel="exp(-x)",
    # yscale="log",
)
ax.plot(Ns, results, "-k")
ax.legend()
ax.set_title('x = 1.0')
```

```python
true_ans = np.full(100, np.exp(-x))
relative_err = (results - true_ans)/true_ans

fig, ax = plt.subplots()
ax.set(
    xlabel="Number of terms N",
    ylabel="log10(abs(relative error))",
    # yscale="log",
)
ax.plot(Ns, np.log10(np.abs(relative_err)), "-k")
ax.legend()
ax.set_title('x = 1.0')
```

### Method 2

```python
def exp_neg_method2(x, N):

    previous_term = 1
    value = 1 # add the 0th term beforehand

    for i in range(1, N+1):
        current_term = previous_term*(-x/i)
        value = value + current_term
        # Update
        previous_term = 1*current_term
    
    return value
```

```python
exp_neg_method2(1,100)
```

```python
results2 = np.zeros(100)

for N in Ns:
    results2[N-1] = exp_neg_method2(x, N)

relative_err2 = (results2 - true_ans)/true_ans
```

```python
fig, ax = plt.subplots()
ax.set(
    xlabel="Number of terms N",
    ylabel="log10(abs(relative error))",
)
ax.plot(Ns, np.log10(np.abs(relative_err)),  "-k", label='Method 1')
ax.plot(Ns, np.log10(np.abs(relative_err2)), "-b", label='Method 2')
ax.legend()
ax.set_title('x = 1.0')
```

### Method 3

```python
def exp_method3(x, N):

    previous_term = 1
    value = 1 # add the 0th term beforehand

    for i in range(1, N+1):
        current_term = previous_term*(x/i)
        value = value + current_term
        # Update
        previous_term = 1*current_term
    
    return value
```

#### x=1.0

```python
x = 1.0
results3 = np.zeros(100)

for N in Ns:
    results3[N-1] = 1/exp_method3(x, N)

relative_err3 = (results3 - true_ans)/true_ans
```

```python
fig, ax = plt.subplots()
ax.set(
    xlabel="Number of terms N",
    ylabel="log10(abs(relative error))",
)
ax.plot(Ns, np.log10(np.abs(relative_err)),  "-k", label='Method 1')
ax.plot(Ns, np.log10(np.abs(relative_err2)), "-b", label='Method 2')
ax.plot(Ns, np.log10(np.abs(relative_err3)), "-r", label='Method 3')
ax.legend()
ax.set_title('x = 1.0')
```
