```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bisection = np.loadtxt('bisection.txt').transpose()
newton    = np.loadtxt('newton.txt').transpose()
secant    = np.loadtxt('secant.txt').transpose()
```

```python
fig, ax = plt.subplots()

ax.set(
    xlabel='N',
    ylabel='$log(error)$'
)
ax.plot(bisection[0], np.log10(bisection[2]), label='Bisectoin')
ax.plot(newton[0],    np.log10(newton[2]),    label='Newton')
ax.plot(secant[0],    np.log10(secant[2]),    label='Secant')

ax.legend()
```
