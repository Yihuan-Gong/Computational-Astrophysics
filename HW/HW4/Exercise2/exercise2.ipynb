```python
import numpy as np
import matplotlib.pyplot as plt
```

### Binary trajectory simulation

```python
def plot_trajectory(Star1, Star2, title='Euler method dt = 0.01 yr'):
    fig, ax = plt.subplots(figsize=(5,5))
    ax.set(
        xlabel='x (m)',
        ylabel='y (m)',
        title=title
    )
    ax.plot(Star1[3], Star1[4], label='Star1')
    ax.plot(Star2[3], Star2[4], label='Star2')
    ax.legend()
```

#### Euler method

```python
# ###########################################
# Read the result from Fotran program

# FORMAT:
# data = np.loadtxt(<filename>).transpose()
# data[1] : Time
# data[3] : Position in x
# data[4] : Position in y
# #############################################

Star1    = np.loadtxt('./outputs/Euler/dt1e-5yr/binary_001.dat').transpose()
Star2    = np.loadtxt('./outputs/Euler/dt1e-5yr/binary_002.dat').transpose()
```

```python
plot_trajectory(Star1, Star2, title='Euler method dt = 0.00001 yr')
```

```python
Star1    = np.loadtxt('./outputs/Euler/dt1e-2yr/binary_001.dat').transpose()
Star2    = np.loadtxt('./outputs/Euler/dt1e-2yr/binary_002.dat').transpose()
```

```python
plot_trajectory(Star1, Star2, title='Euler method dt = 0.01 yr')
```
