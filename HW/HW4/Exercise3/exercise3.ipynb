```python
import numpy as np
import matplotlib.pyplot as plt
```

### Angry bird trajectory simulation

```python
# ###########################################
# Read the result from Fotran program

# FORMAT:
# data = np.loadtxt(<filename>).transpose()
# data[0] : Time
# data[1] : Position in x
# data[2] : Position in y
# #############################################

RK4      = np.loadtxt('./RK4_trajectory.txt').transpose()
```

```python
# Analytical solution
v     = 30
angle = np.deg2rad(16.484375)
vx    =  v*np.cos(angle)
vy    =  v*np.sin(angle)
g     =  9.80

time      = np.linspace(0,1)
posx_anal = vx*time
posy_anal = vy*time - 0.5*g*time**2
```

```python
fig, ax = plt.subplots()

ax.set(
    xlabel='x (m)',
    ylabel='y (m)',
    xlim = [0,50],
    ylim = [0,4]
)

ax.plot(posx_anal, posy_anal, label='Analytical')
ax.plot(RK4[1],   RK4[2],   label='RK4 method')

ax.legend()
```
