```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Calculate area of unit circle by Monte Carlo

```python
def MC_circle(N):

    # Random seed for x and y
    seedx = 100
    seedy = 200

    # Gnerate independent random series for x and y
    np.random.seed(seedx)
    x = np.random.random(N)

    np.random.seed(seedy)
    y = np.random.random(N)

    # Count points within the circle
    r = np.sqrt(x**2 + y**2)
    in_circle = len(r[r<1])

    # Area of the circle
    A = 4*in_circle/N

    return A
```

```python
# Calculate Error for N=10 to 10^7
numbers = np.logspace(1,7)
results = []

for number in numbers:
    N = int(number)
    results.append(MC_circle(N))

results = np.array(results)
errors = np.abs((results - np.pi)/np.pi)
```

```python
fig, ax = plt.subplots()
ax.set(
    xlabel='log(N)',
    ylabel='$log(\epsilon)$',
    # xscale='log',
    # yscale='log',
    title='Monte Carlo unit circle'
)
ax.plot(np.log10(numbers), np.log10(errors))
ax.plot(np.log10(numbers), np.log10(errors[0]*(numbers/numbers[0])**(-0.5)), '--k')
```

#### Calculate integration of $x^2$ form x=0 to 2 by Monte Carlo

```python
def MC_integrate(N):

    # Random seed for x and y
    seedx = 100
    seedy = 200

    # bounds
    x_max = 2
    y_max = 4

    # Gnerate independent random series for x and y
    np.random.seed(seedx)
    x = np.random.random(N)*x_max

    np.random.seed(seedy)
    y = np.random.random(N)*y_max

    # f(x) = f = x^2
    f = x**2
    under_f = len(y[y<f])

    # Area of the circle
    A = x_max*y_max*under_f/N

    return A
```

```python
# Calculate Error for N=10 to 10^7
numbers = np.logspace(1,7)
results = []

for number in numbers:
    N = int(number)
    results.append(MC_integrate(N))

results = np.array(results)
errors = np.abs((results - 8/3)/(8/3))
```

```python
fig, ax = plt.subplots()
ax.set(
    xlabel='log(N)',
    ylabel='$log(\epsilon)$',
    # xscale='log',
    # yscale='log',
    title='Monte Carlo integrate over $x^2$'
)
ax.plot(np.log10(numbers), np.log10(errors))
ax.plot(np.log10(numbers), np.log10(errors[0]*(numbers/numbers[0])**(-0.5)), '--k')
```
