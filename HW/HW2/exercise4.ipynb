```python
import numpy as np
import matplotlib.pyplot as plt
```

### Midpoint

```python
# 32 bit
numbers = np.logspace(1,7,7)
errors = np.abs(np.loadtxt('pi_error.dat'))
```

```python
fig, ax = plt.subplots()
ax.set(
    xlabel='log(N)',
    ylabel='$log(\epsilon)$',
    # xscale='log',
    # yscale='log',
    # ylim=[10**(-7),10**(-1)],
    ylim=[-7, -1],
    title='32 bit Midpoint'
)
ax.plot(np.log10(numbers), np.log10(errors), label = 'relative error')
ax.plot(np.log10(numbers), np.log10(errors[0]*(numbers/numbers[0])**(-2)),
        '--k', label='$1/N^2$')
ax.legend()
```
