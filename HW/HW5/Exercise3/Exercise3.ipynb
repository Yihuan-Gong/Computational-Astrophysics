```python
def get_dt(
        c    = 1. ,
        xmin = 0. ,
        xmax = 1. ,
        imax = 50 ,
        CFL  = 0.4,
    ):

    
    dx  = (xmax-xmin)/imax
    dt  = CFL*dx/c

    return dt
```
