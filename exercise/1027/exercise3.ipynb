```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# By Euler method
euler      = np.loadtxt('./angry_bird/angry_euler.txt').transpose()
time       = euler[0]
posx_euler = euler[1]
posy_euler = euler[2]
```

```python
# Analytical solution
v     = 30
angle = np.deg2rad(60)
vx    =  v*np.cos(angle)
vy    =  v*np.sin(angle)
g     =  9.81

time      = np.linspace(0,7)
posx_anal = vx*time
posy_anal = vy*time - 0.5*g*time**2
```

```python
fig, ax = plt.subplots()

ax.set(
    xlabel='x (m)',
    ylabel='y (m)',
    ylim=[0,50]
)
ax.plot(posx_anal,  posy_anal,  label='Analytical')
ax.plot(posx_euler, posy_euler, label='Euler method')

ax.legend()
```
