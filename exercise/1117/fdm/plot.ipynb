```python
import numpy as np
import matplotlib.pyplot as plt
```

### Upwind

```python
data = np.loadtxt('upwind/advection_00000.d').transpose()
plt.plot(data[1], data[2], label='n=0')

for i in range(100, 800, 100):
    data = np.loadtxt('upwind/advection_00%d.d'%i).transpose()
    plt.plot(data[1], data[2], label='n=%d'%i)

plt.xlabel('x');
plt.ylabel('u(x)');
plt.title('Upwind method CFL=0.4')
plt.legend()
```

### FTCS (Foward Time Central Space)

```python
data = np.loadtxt('advection_00000.d').transpose()
plt.plot(data[1], data[2], label='n=0')

for i in range(10, 50, 10):
    data = np.loadtxt('advection_000%d.d'%i).transpose()
    plt.plot(data[1], data[2], label='n=%d'%i)

plt.xlabel('x');
plt.ylabel('u(x)');
plt.title('FTCS method CFL=0.4')
plt.legend()
```
